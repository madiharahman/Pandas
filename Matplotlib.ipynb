2. What is a scatter plot? Use the following code to generate data for x and y. 
Using this generated data plot a scatter plot.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
#Note: Also add title, xlabel, and ylabel to the plot.
```

`Scatter plots are the graphs that present the relationship between two variables in a data-set. It represents data points on a two-dimensional plane or on a Cartesian system. The independent variable or attribute is plotted on the X-axis, while the dependent variable is plotted on the Y-axis.`

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("TITLE")
```

3. Why is the subplot() function used? Draw four line plots using the subplot() function.

Use the following data:

```python
import numpy as np

#For line 1: 
x1 = np.array([0, 1, 2, 3, 4, 5]) #and 
y1 = np.array([0, 100, 200, 300, 400, 500])

#For line 2: 
x2 = np.array([0, 1, 2, 3, 4, 5]) #and 
y2 = np.array([50, 20, 40, 20, 60, 70])

#For line 3: 
x3 = np.array([0, 1, 2, 3, 4, 5]) #and 
y3 = np.array([10, 20, 30, 40, 50, 60])

#For line 4: 
x4 = np.array([0, 1, 2, 3, 4, 5]) #and 
y4 = np.array([200, 350, 250, 550, 450, 150])
```

`The subplot() function in Matplotlib is used to create multiple plots in a single figure. It takes three arguments: the number of rows, the number of columns, and the plot number. The plot number starts at 1 and increases from left to right and from top to bottom.`

```python
f, ax = plt.subplots(2,2)
ax[0,0].plot(x1, y1)
ax[0,1].plot(x2, y2)
ax[1,0].plot(x3, y3)
ax[1,1].plot(x4, y4)
plt.show()
```

4. What is a bar plot? Why is it used? 
Using the following data plot a bar plot and a horizontal bar plot.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

`A bar plot is a type of plot that represents categorical data with rectangular bars. The height or length of each bar corresponds to the value of a particular category.

Bar plots are commonly used to compare the relative sizes of different categories or to show changes in a categorical variable over time. They are useful for displaying discrete data, such as counts or percentages.`

```python
plt.bar(company,profit)
```

```python
plt.barh(company,profit)
```

5. What is a box plot? Why is it used? 
Using the following data plot a box plot.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

`A box and whisker plotalso called a box plotdisplays the five-number summary of a set of data. The five-number summary is the minimum, first quartile, median, third quartile, and maximum. In a box plot, we draw a box from the first quartile to the third quartile. A vertical line goes through the box at the median.

Box plots provide a quick visual summary of the variability of values in a dataset. Outliers can reveal mistakes or unusual occurrences in data.`

```python
data = [box1, box2]
plt.boxplot(data)
plt.show()
```
