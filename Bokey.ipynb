2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.

`Usually, a plot consists of geometric shapes either in the form of a line, circle, etc. So, Glyphs are nothing but visual shapes that are drawn to represent the data such as circles, squares, lines, rectangles, etc.`  

```python
#To add glyphs to a Bokeh plot, we can use the glyph methods provided by the figure class.
x = [2,3,4,5,6]
y = [3,4,5,6,7]
output_file("line.html")
p = figure(title = "line plot")
p.line(x,y)
show(p)
```

4. What is a Bokeh server, and how can you use it to create interactive plots that can be updated in real time?

`Bokeh server makes it easy to create interactive web applications that connect front-end UI events to running Python code. The Bokeh server can handle multiple connections simultaneously and can generate dynamic content that updates automatically as users interact with the plot.`

`To use the Bokeh server, you need to define a Bokeh curdoc object that contains the plot and any widgets or tools that you want to include. We can then launch the Bokeh server using the bokeh serve command, specifying the file path to your script or directory.`

5. How can you embed a Bokeh plot into a web page or dashboard using Flask or Django?

**Flask:**

Standalone HTML File:
* Create a Bokeh plot.
* Use Bokeh's save function to generate an HTML file.
* Render the HTML file within a Flask route.

Bokeh Server:
* Install Bokeh server.
* Define a Bokeh server app using the modify_doc function.
* Run the Bokeh server.
* Embed the Bokeh server script into a Flask template.

**Django:**

Standalone HTML File:
* Similar to Flask, generate a standalone HTML file and render it within a Django view.

Bokeh Server:
* Embed the Bokeh server script into a Django template.
* Create a Django view to serve the Bokeh server script.
